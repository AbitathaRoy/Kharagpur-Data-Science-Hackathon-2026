```python
import pandas as pd
```

```python
train_csv = pd.read_csv("../data/train.csv")
train_csv.columns
```

```python
print(f"Unique Books: {train_csv['book_name'].value_counts()}")
print(f"\nUnique Characters: {train_csv['char'].value_counts()}")
```

```python
# Caption and content length distribution
train_csv['caption_length'] = train_csv['caption'].astype(str).str.len()
train_csv['content_length'] = train_csv['content'].astype(str).str.len()

# If you specifically need them as Python lists:
caption_length_list = train_csv['caption_length'].tolist()
content_length_list = train_csv['content_length'].tolist()
```

```python
import matplotlib.pyplot as plt 
from collections import Counter

# Count frequencies 
freq1 = Counter(caption_length_list) 
freq2 = Counter(content_length_list)

# Sort values for consistent plotting
x1 = sorted(freq1.keys())
y1 = [freq1[val] for val in x1]

x2 = sorted(freq2.keys())
y2 = [freq2[val] for val in x2]

# Plot line plots
plt.figure(figsize=(20, 6))
plt.plot(y1, x1, marker='o', label='Caption')
plt.plot(y2, x2, marker='s', label='Content')

# Labels and title
plt.xlabel("Frequency")
plt.ylabel("Numerical Value")
plt.title("Frequency Line Plot of Two Lists")
plt.legend()
plt.grid(True)
plt.show()
```
